```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('all_data.csv')
df.rename(columns={'Life expectancy at birth (years)': 'life_expectancy'}, inplace=True)
print(df.head(60))
countries = df.Country.unique()
print(countries)
print(df[df.Country == 'Zimbabwe'])
```

```python
plt.figure(figsize=(12,6))
sns.lmplot(data=df, y='life_expectancy', x='GDP', x_jitter=0.05e13)
plt.show()
plt.clf()
```

```python
for country in countries:
    ax=sns.lmplot(data=df[df.Country == country], y='life_expectancy', x='GDP', hue='Year', palette='flare', line_kws={'color': 'black'}, fit_reg=True)
    ax.set(title=f'{country} Life Expectancy vs GDP',
           xlabel='GDP (USD)',
           ylabel = 'Life Expectancy at Birth (Years)')
    plt.show()
    plt.clf()
```

```python
avg_life_exp = [np.mean(df[df.Country == country].life_expectancy) for country in countries]
print(avg_life_exp)
avg_gdp = [np.mean(df[df.Country == country].GDP) for country in countries]
print(avg_gdp)
for i in range(len(countries)):
    print(f'From 2000 to 2015, {countries[i]} had an average GDP of {"{:e}".format(avg_gdp[i])} USD and an average life expectancy (at birth) of {round(avg_life_exp[i],2)} years.')
```

## Has life expectancy increased over time in the six nations?
    Yes. There is a distinct increase in life expectancy for each nation from 2000 to 2015, with only Zimbabwe showing some decrease in life expectancy in 2001-2004. 
    
## Has GDP increased over time in the six nations?
    Yes. There is also a distinct increase in GDP for each nation from 2000 to 2015, with Zimbabwe also showing setbacks to their GDP in 2001-2004.

## Is there a correlation between GDP and life expectancy of a country?
        Yes. Both life expectancy and GDP increased distinctly from 2000 to 2015. Zimbabwe suffered an economic crisis* in the early 2000's, where a temporary decrease in GDP correlated with a temporary decrease in life expectancy. 
        Zimbabwe's GDP is distinctly lower than the other countries' GDP's, so their relatively low life expectancy further supports a correlation between GDP and life expectancy.

## What is the average life expectancy in these nations?
    Zimbabwe had an average life expectancy of 50 years from 2000 to 2015, whereas the others averaged to 75-80 years. 

## What is the distribution of that life expectancy?
    The distribution is centered around 75-80, but is highly left-skewed due to Zimbabwe.
